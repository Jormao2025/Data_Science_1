```python
# Importar librerías
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
```

```python
# Configuraciones opcionales para mejor visualización
sns.set_theme(style="whitegrid") # Establece un tema estético para los gráficos
plt.rcParams['figure.figsize'] = (12, 6) # Establece un tamaño de figura por defecto
warnings.filterwarnings('ignore') # Ignora advertencias (útil para presentaciones)
```

```python
import requests

url = "https://github.com/Jormao2025/Data_Science_1/raw/refs/heads/main/Dataset%20Proyecto%20Coderhouse%20Satisfacci%C3%B3n%20Clientes%20de%20Aerol%C3%ADneas.csv"

response = requests.get(url)  # Hace una solicitud GET para obtener el contenido del archivo
print(response.text[:300])  # Imprime los primeros 300 caracteres del archivo para ver cómo está delimitado
response = requests.get(url)  # Hace una solicitud GET para obtener el contenido del archivo
print(response.text[:300])  # Imprime los primeros 300 caracteres del archivo para ver cómo está delimitado
```

```python
df = pd.read_csv(url, sep=';', encoding='latin1') # lee el archivo,

df.head()# Muestra las primeras 5 filas
```

```python
df.describe()  # Estadísticas básicas de columnas numéricas
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
df.isnull().sum()  # Cuenta valores faltantes por columna
```

```python
df["Retraso en la llegada en minutos"].fillna("No", inplace=True)##rellena los vacios con "No_encontrado"
```

```python
df.isnull().sum()  # Cuenta valores faltantes por columna
```

```python
df.info()      # muestra tipos de datos y valores no nulos
```

```python
df['Retraso en la llegada en minutos'] = pd.to_numeric(df['Retraso en la llegada en minutos'], errors='coerce') #verificamos valores con error
```

```python
df['Retraso en la llegada en minutos'].isna().sum() # transformamos los valores inválidos
```

```python
df['Retraso en la llegada en minutos'].fillna(0, inplace=True) #Revisamos si hay NaN
```

```python
df['Retraso en la llegada en minutos'] = df['Retraso en la llegada en minutos'].astype('int64') #rellenamos con 0 los NaN
```

```python
df.info()      # muestra tipos de datos con los cambios realizados
```

```python
# Gráfico de Barras para conocer la Satisfacción de Clientas de Aerolíneas
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, x='Género', hue='Satisfacción', palette='Set2')
plt.title('Satisfacción de Clientes de Aerolíneas por Género')
plt.xlabel('Género')
plt.ylabel('Número de Clientes')
plt.legend(title='Satisfacción')
plt.grid(False)

# Agregar etiquetas en cada barra
for p in ax.patches:
    altura = p.get_height()
    ax.annotate(f'{int(altura)}',  # convierte a entero
                (p.get_x() + p.get_width() / 2., altura),  # posición
                ha='center', va='bottom', fontsize=8, color='grey', xytext=(0, 5), textcoords='offset points')

# Ajustar la escala del eje y
plt.ylim(0, 10000)

plt.show()
```

```python
# Agrupamos por Género y Satisfacción
genero_satisfaccion = df.groupby(["Género", "Satisfacción"]).size().unstack()

# Creamos los gráficos de torta lado a lado
fig, axes = plt.subplots(1, 2, figsize=(10,5))

colores = ["#e89675" , "#72b5a0"]

# Pie chart para Femenino
genero_satisfaccion.loc["Femenino"].plot(
    kind="pie", autopct="%.1f%%", ax=axes[0], colors=colores, startangle=90
)
axes[0].set_ylabel("")
axes[0].set_title("Femenino")

# Pie chart para Masculino
genero_satisfaccion.loc["Masculino"].plot(
    kind="pie", autopct="%.1f%%", ax=axes[1], colors=colores, startangle=90
)
axes[1].set_ylabel("")
axes[1].set_title("Masculino")

plt.suptitle("Proporción de Satisfacción de Clientes de Aerolíneas", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras sobre Satisfacción de Clientes en Aereolíneas según Tipo de Viaje
sns.set_style("white")
plt.figure(figsize=(10, 5))
plt.title('Satisfacción del Cliente según Tipo de Viaje')

# Agregar etiquetas de datos
ax = sns.countplot(data=df, x='Tipo de Viaje', hue='Satisfacción', palette='Set2')

# Agregar etiquetas en cada barra
for p in ax.patches:
    altura = p.get_height()
    ax.annotate(f'{int(altura)}',  # convierte a entero
                (p.get_x() + p.get_width() / 2., altura),  # posición
                ha='center', va='bottom', fontsize=8, color='grey', xytext=(0, 5), textcoords='offset points')

  # Ajustar la escala del eje y para que las etiquetas no se corten
plt.ylim(0, 12000)
plt.show()
```

```python
# Boxplot: Distribución de la satisfacción por edad
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Satisfacción", y="Edad", palette="Set2")
plt.title("Distribución de Edad por Nivel de Satisfacción")
plt.xlabel("Satisfacción")
plt.ylabel("Edad")
plt.show()
```

```python
# Displot: Distribución de la distiancia de los vuelos según la satisfaccion del pasajero
sns.displot(data=df, x='Distancia de Vuelo', col='Satisfacción', kde=True, height=4, aspect=1.2)
plt.suptitle('Distribución de la distancia de los vuelos según la satisfacción del pasajero', y=1.05) # Título general ajustado
plt.show()
```

```python
# Filtrar el DataFrame para incluir solo los pasajeros insatisfechos
df_insatisfechos = df[df['Satisfacción'] == 'neutral o insatisfecho/a']

# Gráfico de violín para analizar la distribución de los insatisfechos por tipo de viaje
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Satisfacción', hue='Tipo de Viaje', palette='Set2', split=True)

# Títulos y etiquetas
plt.title('Distribución de Insatisfacción por Distancia de Vuelo y Tipo de Viaje')
plt.xlabel('Distancia de Vuelo')
plt.ylabel('Satisfacción')

plt.show()
```

```python
#Gráfico de promedio por motivo y satisfacción
motivos = ['Hora de salida/llegada conveniente', 'Facilidad de reserva en línea', 'Ubicación de la puerta','Comida y bebida','Embarque en línea', 'Comodidad del asiento', 'Entretenimiento a bordo', 'Servicio a bordo', 'Servicio de espacio para las piernas', 'Manejo de equipaje', 'Servicio de facturación', 'Servicio a bordo.1' , 'Limpieza' ]
#defino todos los motivos de puntuación del 1 al 5 bajo "motivos"
```

```python
#Gráfico de promedio por motivo y satisfacción
df.melt(id_vars=['Satisfacción'], value_vars=motivos,
        var_name='motivos', value_name='Puntuacion')
# Transformo el DataFrame a formato largo: dejo 'Satisfacción' fija y convierto los motivos en filas
```

```python
df['Satisfacción'] = df['Satisfacción'].map({'neutral o insatisfecho/a': 0, 'satisfecho/a': 1})
```

```python
# Mapa de calor de correlaciones
import numpy as np
plt.figure(figsize=(15, 10))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlaciones entre variables numéricas")
plt.show()
```

```python
# Filtrar el DataFrame para incluir solo los pasajeros insatisfechos
df_insatisfechos = df[df['Satisfacción'] == 0]

# Crear el gráfico relplot using relevant columns
sns.relplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Servicio a bordo', hue='Tipo de Viaje', col='Tipo de Viaje',
kind='scatter', height=5, aspect=0.9, s=70, alpha=0.8, palette='viridis')

plt.suptitle('Relación entre Distancia de Vuelo y Servicio a Bordo (Separado por Tipo de Viaje)', y=1.03, fontsize=16)

# Ajustar etiquetas individualmente si es necesario
plt.gcf().axes[0].set_xlabel("Distancia de Vuelo")
plt.gcf().axes[0].set_ylabel("Servicio a Bordo Rating")

plt.tight_layout(rect=[0, 0, 1, 0.97]) # Ajustar para que el suptitle no se solape
plt.show()
```

```python
# Filtrar el DataFrame para incluir solo los pasajeros insatisfechos
df_insatisfechos = df[df['Satisfacción'] == 0] # Filter using the numerical value 0

# Crear el gráfico relplot
sns.relplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Retraso en la llegada en minutos',
hue='Tipo de Viaje', col='Tipo de Viaje',
kind='scatter', height=5, aspect=0.9, s=70, alpha=0.8, palette='viridis')

plt.suptitle('Relación entre Distancia de Vuelo y Retraso (Separado por Tipo de Viaje)', y=1.03, fontsize=16)

# Ajustar etiquetas individualmente si es necesario
plt.gcf().axes[0].set_xlabel("Distancia de Vuelo")
plt.gcf().axes[0].set_ylabel("Retraso en la llegada en minutos")

plt.tight_layout(rect=[0, 0, 1, 0.97]) # Ajustar para que el suptitle no se solape
plt.show()
```

```python
# Para una visión más amplia de la insatisfacción:
# Servicio a bordo vs Comodidad del asiento (crucial para viajes largos)
sns.relplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Comodidad del asiento',
            hue='Tipo de Viaje', col='Tipo de Viaje', kind='scatter',
            height=5, aspect=0.9, s=70, alpha=0.8, palette='viridis')

plt.suptitle('Relación entre Distancia de Vuelo y Comodidad del Asiento', y=1.03, fontsize=16)
```

```python
# Para analizar servicios digitales:
# Servicio de wifi vs Distancia (importante para viajeros de negocios)
sns.relplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Servicio de wifi a bordo',
            hue='Clase', col='Tipo de Viaje', kind='scatter',
            height=5, aspect=0.9, s=70, alpha=0.8, palette='viridis')

plt.suptitle('WiFi a Bordo vs Distancia de Vuelo por Tipo de Viaje y Clase', y=1.03, fontsize=16)
```

```python
# Para experiencia integral del viaje: Combinación de servicios clave
sns.relplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Entretenimiento a bordo',
            size='Comida y bebida', hue='Servicio a bordo', col='Tipo de Viaje',
            kind='scatter', height=5, aspect=0.9, alpha=0.7, palette='viridis')

plt.suptitle('Experiencia Integral vs Distancia por Tipo de Viaje', y=1.03, fontsize=16)
```

```python
# Análisis por clase y tipo de cliente: Servicio a bordo vs Distancia, segmentado por clase y tipo de cliente
sns.relplot(data=df_insatisfechos, x='Distancia de Vuelo', y='Servicio a bordo',
            hue='Clase', col='Tipo de Cliente', row='Tipo de Viaje',
            kind='scatter', height=4, aspect=1.0, s=60, alpha=0.8, palette='viridis')

plt.suptitle('Servicio a Bordo por Distancia, Clase y Tipo de Cliente', y=1.03, fontsize=16)
```

```python
# Para identificar problemas operativos: Relación con retrasos (puede ser clave en insatisfacción)
sns.relplot(data=df_insatisfechos, x='Retraso en la salida en minutos', y='Servicio a bordo',
hue='Distancia de Vuelo', col='Tipo de Viaje',
kind='scatter', height=5, aspect=0.9, s=70, alpha=0.8, palette='viridis')

plt.suptitle('Retrasos vs Servicio a Bordo por Tipo de Viaje', y=1.03, fontsize=16)
```

```python
# Para Analizar la Experiencia Digital y de Reserva (Crítico para viajeros de negocios) Estos datos revelan si los pasajeros que encuentran fácil reservar online luego califican mal el WiFi, hay una desconexión en la experiencia digital integral.
# 1.1 Facilidad de Reserva Online vs Servicio de WiFi (¿La promesa digital se cumple?)
sns.relplot(data=df_insatisfechos, x='Facilidad de reserva en línea', y='Servicio de wifi a bordo',
            hue='Tipo de Viaje', col='Tipo de Viaje', kind='scatter', alpha=0.8)
plt.suptitle('Facilidad de Reserva vs Calidad de WiFi (Por Tipo de Viaje)', y=1.03)

# 1.2 Embarque en línea vs Manejo de equipaje (¿Agiliza el proceso total?)
sns.relplot(data=df_insatisfechos, x='Embarque en línea', y='Manejo de equipaje',
            hue='Tipo de Cliente', style='Clase', kind='scatter', alpha=0.7)
plt.suptitle('Embarque Online vs Manejo de Equipaje (Por Tipo de Cliente y Clase)', y=1.03)
```

```python
# Para Analizar la Experiencia en Cabina (Confort físico)
# 2.1 Comodidad del asiento vs Servicio de espacio para las piernas (Variables obviamente correlacionadas)
sns.relplot(data=df_insatisfechos, x='Comodidad del asiento', y='Servicio de espacio para las piernas',
            hue='Clase', size='Distancia de Vuelo', kind='scatter', alpha=0.7)
plt.suptitle('Relación entre Comodidad y Espacio para Piernas (Tamaño por Distancia)', y=1.03)

# 2.2 Comida y bebida vs Entretenimiento a bordo (Los dos "servicios de entretenimiento")
sns.relplot(data=df_insatisfechos, x='Comida y bebida', y='Entretenimiento a bordo',
            col='Tipo de Viaje', hue='Clase', kind='scatter', alpha=0.8)
plt.suptitle('Comida/Bebida vs Entretenimiento (Por Tipo de Viaje y Clase)', y=1.03)
```

```python
# Para Analizar el Impacto de los Retrasos (Factor operacional clave)
# 3.1 Retraso en salida vs Servicio a bordo (¿El mal humor por el retraso afecta la percepción del servicio?)
sns.relplot(data=df_insatisfechos, x='Retraso en la salida en minutos', y='Servicio a bordo',
            hue='Tipo de Viaje', col='Tipo de Viaje', kind='scatter', alpha=0.8)
plt.suptitle('Impacto del Retraso en Salida sobre la Percepción del Servicio', y=1.03)

# 3.2 Retraso en salida vs Comida y bebida (¿Los retrasos afectan la calidad de la comida?)
sns.relplot(data=df_insatisfechos, x='Retraso en la salida en minutos', y='Comida y bebida',
            hue='Distancia de Vuelo', kind='scatter', alpha=0.7, palette='viridis')
plt.suptitle('Retrasos vs Calidad de Comida/Bebida (Color por Distancia)', y=1.03)
```

```python
# Para Analizar el Perfil del Cliente Insatisfecho
# 4.1 Edad vs Servicio a bordo (¿Hay un grupo etario más crítico?)
sns.relplot(data=df_insatisfechos, x='Edad', y='Servicio a bordo',
            hue='Tipo de Viaje', style='Género', kind='scatter', alpha=0.7)
plt.suptitle('Edad vs Evaluación del Servicio (Por Tipo de Viaje y Género)', y=1.03)

# 4.2 Edad vs Comodidad del asiento (¿Los pasajeros mayores son más sensibles a la comodidad?)
sns.relplot(data=df_insatisfechos, x='Edad', y='Comodidad del asiento',
            hue='Clase', kind='scatter', alpha=0.8)
plt.suptitle('Edad vs Comodidad del Asiento (Por Clase)', y=1.03)
```

```python
# Para Identificar las Causas Principales de la Insatisfacción
# Filtrar solo las columnas numéricas de servicio/valoración
columnas_de_servicio = ['Servicio de wifi a bordo', 'Facilidad de reserva en línea', 'Comida y bebida',
                        'Embarque en línea', 'Comodidad del asiento', 'Entretenimiento a bordo',
                        'Servicio a bordo', 'Servicio de espacio para las piernas', 'Manejo de equipaje',
                        'Servicio de facturación', 'Limpieza']

# Calcular la matriz de correlación
correlacion = df_insatisfechos[columnas_de_servicio].corr()

# Crear el heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlación entre Servicios para Pasajeros Insatisfechos\n(Valores cerca de +1: fuertemente relacionados)')
plt.tight_layout()
plt.show()
```

```python
# Para Comparar el Desempeño por Grupo (Tipo de Viaje, Clase)
# Crear una columna combinada para segmentar mejor
df_insatisfechos['Segmento'] = df_insatisfechos['Tipo de Viaje'] + ' - ' + df_insatisfechos['Clase']

plt.figure(figsize=(14, 6))
sns.boxplot(data=df_insatisfechos, x='Segmento', y='Servicio a bordo', hue='Tipo de Cliente')
plt.title('Distribución de las Calificaciones del Servicio a Bordo por Segmento del Pasajero')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Para Entender la Composición de los Insatisfechos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: Tipo de Viaje
sns.countplot(data=df_insatisfechos, x='Tipo de Viaje', ax=ax1, palette='viridis')
ax1.set_title('Número de Pasajeros Insatisfechos por Tipo de Viaje')

# Gráfico 2: Clase
sns.countplot(data=df_insatisfechos, x='Clase', ax=ax2, palette='viridis')
ax2.set_title('Número de Pasajeros Insatisfechos por Clase')

plt.tight_layout()
plt.show()
```

```python
# Para Analizar el Impacto de los Retrasos
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_insatisfechos,
                x='Retraso en la salida en minutos',
                y='Retraso en la llegada en minutos',
                alpha=0.5,  # Hace los puntos semi-transparentes para ver concentración
                color='red',
                s=30)       # Tamaño de los puntos

plt.title('Relación entre Retrasos en Salida y Llegada para Pasajeros Insatisfechos')
plt.xlabel('Retraso en Salida (minutos)')
plt.ylabel('Retraso en Llegada (minutos)')
plt.tight_layout()
plt.show()
```
